```python
import pandas as pd
import numpy as np

year_used = "2016"
columns = ["State",year_used]
```

```python
#Political Factor DataFrame (based on Popular Vote in 2016 Presidental Election)
election2016_file = "Resources/2016/data/election2016.csv"
election2016_data =  pd.read_csv(election2016_file)
election2016_data = pd.DataFrame(election2016_data)
gov_election2016_data = election2016_data.copy()
gov_election2016_data = gov_election2016_data.iloc[:56,:4] 
convert_to_number_list = ["Democratic Candidates","Republican Candidates","Other Candidates"]
for item in ["Democratic Candidates","Republican Candidates","Other Candidates"]:
    for x in range(len(gov_election2016_data[item])):
        gov_election2016_data[item][x] = int(gov_election2016_data[item][x].replace(',' , ''))
    
us_states = {
        'AK': 'Alaska',
        'AL': 'Alabama',
        'AR': 'Arkansas',
        'AS': 'American Samoa',
        'AZ': 'Arizona',
        'CA': 'California',
        'CO': 'Colorado',
        'CT': 'Connecticut',
        'DC': 'District of Columbia',
        'DE': 'Delaware',
        'FL': 'Florida',
        'GA': 'Georgia',
        'GU': 'Guam',
        'HI': 'Hawaii',
        'IA': 'Iowa',
        'ID': 'Idaho',
        'IL': 'Illinois',
        'IN': 'Indiana',
        'KS': 'Kansas',
        'KY': 'Kentucky',
        'LA': 'Louisiana',
        'MA': 'Massachusetts',
        'MD': 'Maryland',
        'ME': 'Maine',
        'MI': 'Michigan',
        'MN': 'Minnesota',
        'MO': 'Missouri',
        'MP': 'Northern Mariana Islands',
        'MS': 'Mississippi',
        'MT': 'Montana',
        'NA': 'National',
        'NC': 'North Carolina',
        'ND': 'North Dakota',
        'NE': 'Nebraska',
        'NH': 'New Hampshire',
        'NJ': 'New Jersey',
        'NM': 'New Mexico',
        'NV': 'Nevada',
        'NY': 'New York',
        'OH': 'Ohio',
        'OK': 'Oklahoma',
        'OR': 'Oregon',
        'PA': 'Pennsylvania',
        'PR': 'Puerto Rico',
        'RI': 'Rhode Island',
        'SC': 'South Carolina',
        'SD': 'South Dakota',
        'TN': 'Tennessee',
        'TX': 'Texas',
        'UT': 'Utah',
        'VA': 'Virginia',
        'VI': 'Virgin Islands',
        'VT': 'Vermont',
        'WA': 'Washington',
        'WI': 'Wisconsin',
        'WV': 'West Virginia',
        'WY': 'Wyoming'
}
gov_election2016_data["Party"] = gov_election2016_data["State"]
for state in range(len(gov_election2016_data["State"])):
    try:
        if gov_election2016_data.iloc[state,1] > gov_election2016_data.iloc[state,2]:
            if gov_election2016_data.iloc[state,1] > gov_election2016_data.iloc[state,3]:
                gov_election2016_data.iloc[state, 4] = "D"
        elif gov_election2016_data.iloc[state,2] > gov_election2016_data.iloc[state,1]:
            if gov_election2016_data.iloc[state,2] > gov_election2016_data.iloc[state,3]:
                gov_election2016_data.iloc[state, 4] = "R"
        else:
            gov_election2016_data.iloc[state, 4] = "I"
        gov_election2016_data["State"][state] = us_states[gov_election2016_data["State"][state]]
    except:
        None
gov_election2016_data = gov_election2016_data[["State","Party"]]
gov_election2016_data.to_csv("output/gov_election2016.csv", index=False, header=True)
gov_election2016_data
```

```python
#Population Data Cleanup to DataFrame
file = "Resources/ALL/data/population2010-2019.csv"
total_population_data_file_all = pd.read_csv(file)
total_population_data_file = total_population_data_file_all.copy()
total_population_data_file = total_population_data_file.iloc[5:,4:17]
total_population_data_file = total_population_data_file.reset_index()
total_population_data_file = total_population_data_file.iloc[:,1:]
total_population_data_file = total_population_data_file.rename(columns={"NAME":"State"})
total_population_data_file = total_population_data_file.rename(columns={f"POPESTIMATE{year_used}":f"{year_used}"})
total_population_data_file = total_population_data_file.loc[:,columns]
total_population_data_file = total_population_data_file.iloc[:51]
total_population_data_file.dtypes
total_population_data_file = pd.merge(total_population_data_file,gov_election2016_data, on="State")
total_population_data_file
```

```python
#Veteran Population Data Cleanup to DataFrame
veteran_file = f"Resources/{year_used}/data/veteran_{year_used}.csv"
veteran_data = pd.read_csv(veteran_file)
veteran_data_file = veteran_data.copy()
#S2101_C01_001E Population Data
#S2101_C03_001E Veteran Data
data_columns = ["NAME","S2101_C01_001E","S2101_C03_001E"]
veteran_data_file = veteran_data_file[data_columns]
veteran_data_file = veteran_data_file.iloc[1:,:]
for column in data_columns:
    try:
        veteran_data_file[column] = veteran_data_file[column].astype(np.float64)
    except ValueError:
        None
veteran_data_file[year_used] = veteran_data_file[data_columns[2]]/veteran_data_file[data_columns[1]]
veteran_data_file = veteran_data_file.rename(columns={"NAME":"State"})
veteran_data_file = veteran_data_file[columns]
veteran_data_file = pd.merge(veteran_data_file,gov_election2016_data, on="State")
total_population_data_file
```
