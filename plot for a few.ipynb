```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
%matplotlib notebook
```

```python
#Graphing BarChart for Poverty and Pay Assist 2017

#Uploading File
poverty_file = "Output/pop_poverty.csv"
payassist_file = "Output/mh_payassist_trends.csv"

#Creating DataFrames
poverty_data = pd.read_csv(poverty_file)
payassist_data = pd.read_csv(payassist_file)

poverty_data = pd.DataFrame(poverty_data)
payassist_data = pd.DataFrame(payassist_data)

#Graphing
y = np.arange(len(poverty_data["State"]))

height = 0.5
fig, ax = plt.subplots(figsize=(10,20))
rects1 = ax.barh(y - height/4, poverty_data["2017"], height/2, height, label='Poverty')
rects2 = ax.barh(y + height/4, payassist_data["2017"], height/2, height, label='Pay Assist')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Percentages (%)')
ax.set_title('Poverty and Pay Assist 2017')
ax.set_yticks(y)
ax.set_yticklabels(poverty_data["State"])
ax.invert_yaxis()
ax.legend(loc=0)
plt.savefig("Poverty and Pay Assist 2017.png")
```

```python
x = poverty_data.iloc[:-1]
y = payassist_data.iloc[:-1]

fig, ax = plt.subplots()
ax.scatter(x.iloc[2], y.iloc[2], marker='o')

for x,y in zip(x,y):
    plt.annotate(x.iloc[0], # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center')

ax.set_xlabel("Poverty Percentage (%)")
ax.set_ylabel("Pay Assist Percentage (%)")
plt.show()

```

```python
#Graphing BarChart for 65+ Assist 2017

#Uploading File
pop_file = "Output/pop_age_65andup.csv"
mh_file = "Output/mh_senior_trends.csv"

#Creating DataFrames
pop_data = pd.read_csv(pop_file)
mh_data = pd.read_csv(mh_file)

pop_data = pd.DataFrame(pop_data)
mh_data = pd.DataFrame(mh_data)

#Graphing
y = np.arange(len(pop_data["State"]))
label =pop_data["State"]

height = 0.5
fig, ax = plt.subplots(figsize=(10,20))
rects1 = ax.barh(y - height/4, pop_data["2017"], height/2, height, label='Population')
rects2 = ax.barh(y + height/4, mh_data["2017"], height/2, height, label='MentalHealth')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Percentages (%)')
ax.set_title('Mental Health Usage of Seniors(65+) vs Population of Seniors 2017')
ax.set_yticks(y)
ax.set_yticklabels(label)
ax.invert_yaxis()
ax.legend(loc=0)
plt.savefig("Mental Health Usage of Seniors(65+) vs Population of Seniors 2017.png")
```
