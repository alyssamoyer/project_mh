```python
import pandas as pd
import numpy as np
```

```python
language_file2015 = 'Resources/2015/data/language_2015.csv'
language_file2016 = 'Resources/2016/data/language_2016.csv'
language_file2017 = 'Resources/2017/data/language_2017.csv'
language_file2018 = 'Resources/2018/data/language_2018.csv'

language_2015 = pd.read_csv(language_file2015)
language_2016 = pd.read_csv(language_file2016)
language_2017 = pd.read_csv(language_file2017)
language_2018 = pd.read_csv(language_file2018)

language_2015 = pd.DataFrame(language_2015)
language_2016 = pd.DataFrame(language_2016)
language_2017 = pd.DataFrame(language_2017)
language_2018 = pd.DataFrame(language_2018)

language_datas = [language_2015, language_2016, language_2017, language_2018]
start_year = 2015
language_data_all = language_2015.copy()
language_data_all = language_data_all.loc[1:,["GEO_ID","NAME"]]
language_data_all = language_data_all.rename(columns={"NAME":"State"})

year = start_year
for data in language_datas: 
    data = data[["NAME","S1602_C02_002E","S1602_C02_002M","S1602_C04_002E","S1602_C04_002M"]]
    data = data.rename(columns={"NAME":"State",
                                "S1602_C02_002E":"{} Percentage of Household Spanish Speakers".format(year),
                                "S1602_C02_002M":"{} Margin of Error of Percentage of Household Spanish Speakers".format(year),
                                "S1602_C04_002E":"{} Percentage of Only Spanish Speakers".format(year),
                                "S1602_C04_002M":"{} Margin of Error of Percentage of Only Spanish Speakers".format(year)})
    language_data_all = pd.merge(language_data_all, data, on="State")
    year += 1
language_data_all_final = language_data_all.iloc[:,1:]
language_data_all_final
```
