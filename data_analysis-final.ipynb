```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
%matplotlib inline


#Load tables created from the data cleaning script as datadrames
mh_adolescent_trends = pd.read_csv('Output/mh_adolescent_trends.csv')
mh_language_trends = pd.read_csv('Output/mh_language_trends.csv')
suicide_rate_service = pd.read_csv('Output/suicide_rate_service.csv')
mh_payassist_trends = pd.read_csv('Output/mh_payassist_trends.csv')
mh_lgbt_trends = pd.read_csv('Output/mh_lgbt_trends.csv')
mh_suicide_2017 = pd.read_csv('Output/mh_suicide_2017.csv')
mh_suicide_programs = pd.read_csv('Output/mh_suicide_trends.csv')
mh_suicide_death = pd.read_csv('Output/mh_suicide_all.csv')
mh_cosubstance_programs = pd.read_csv('Output/mh_substance_trends.csv')
mh_veteran_programs = pd.read_csv('Output/mh_verteran_trends.csv')
mh_housing_programs = pd.read_csv('Output/mh_housing_trends.csv')
mh_education_programs = pd.read_csv('Output/mh_education_trends.csv')
```

```python
#Loop through list of states in order of education trends. Create a list of colors. Blue for democrat, red for republican.
party = []
index_count = 0
for rows in mh_education_programs.iterrows(): 
    if mh_education_programs.loc[index_count, 'Party']== 'D':
        party.append('royalblue')
    elif mh_education_programs.loc[index_count, 'Party']== 'R':
        party.append('firebrick')
    else:
        party.append('k') 
    index_count = index_count + 1
    
x_ticks = [i for i, _ in enumerate(mh_education_programs['state'])]

plt.bar(x_ticks, mh_education_programs['2016'], .6, color=party)

plt.ylabel("% of MH Facilities with Education Program")
plt.title("U.S. Mental Health Facilities with Education Programs in 2016 by State and Party Affiliation", wrap = True)

plt.xticks(x_ticks, mh_education_programs['state'], rotation = 90)
plt.style.use('seaborn')

#Create legend for party affiliation
red_patch = mpatches.Patch(color='firebrick', label='Republican')
blue_patch = mpatches.Patch(color='royalblue', label='Democrat')
plt.legend(handles=[red_patch, blue_patch])
plt.tight_layout()
plt.savefig("Images/education_program.png")
plt.show()
```

```python
#Summarize the Education Distribution. Find the mean of the republican/democratic party and the standard deviation of the total distribution
education_d = mh_education_programs.loc[mh_education_programs['Party'] == 'D' ]['2016'].mean()
education_r = mh_education_programs.loc[mh_education_programs['Party'] == 'R' ]['2016'].mean()
education_std = mh_education_programs['2016'].std()
education_r, education_d, education_std
education_df = pd.DataFrame({'Republican Mean':education_r, 
                             'Democrat Mean': education_d, 
                             'Standard Deviation': education_std  }, index=[0])
education_df
```

```python
x_suicide = suicide_rate_service['RATE']
y_suicide = suicide_rate_service['2017']

for i,type in enumerate(suicide_rate_service['state']):
    x = x_suicide[i]
    y = y_suicide[i]
    plt.scatter(x, y, marker='o', color='cornflowerblue', edgecolors='darkblue')
    
    if (x < 10) or (x > 25) or (y < 38) or (y> 80):
        plt.text(x+0.3, y+0.3, type, fontsize=8, wrap=True)

z = np.polyfit(x_suicide, y_suicide, 1)
p = np.poly1d(z)
plt.plot(x_suicide,p(x_suicide),"r--", color = 'dimgray')

plt.xlabel('Suicide Death Rate')
plt.ylabel('% Of MH Facilities Offering Suicide Prevention')
plt.title("2017 Suicide Death Rate vs % of MH Facilities Offering Suicide Prevention Services by State", wrap=True)
plt.grid()
plt.savefig("Images/suicide_rate_service.png")

plt.show()
```
